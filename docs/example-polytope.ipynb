# Access Climate DT data with polytope

```python
import polytope
import xarray as xr
```

```python
rcfile = "~/.polytopeapirc"
```

```python
kwargs = {"address": "polytope.lumi.apps.dte.destination-earth.eu"}
client = polytope.api.Client(**kwargs)
```

```python
client.list_collections()
```

## Request data
- Check [ECMWF Climate DT Overview](https://confluence.ecmwf.int/display/DDCZ/Climate%20DT%20overview) to get information on possible values for each field, e.g. parameters from the ECMWF Parameter Database.

```python
request = {
    "activity": "ScenarioMIP",
    "class": "d1",
    "dataset": "climate-dt",
    "date": "20200102",
    "experiment": "SSP3-7.0",
    "expver": "0001",
    "generation": "1",
    "levtype": "o2d",
    "model": "IFS-NEMO",
    "param": "263100/263101/263124",
    "realization": "1",
    "resolution": "standard",
    "stream": "clte",
    "time": "0000",  # '0100/0200/0300/0400/0500/0600'
    "type": "fc",
}
```

```python
# Retrieve data
client.retrieve("destination-earth", request, "output_api.grib")
```

```python
dset = xr.open_dataset("output_api.grib", engine="cfgrib")
```
