# Access Climate DT data with polytope

```python
import polytope
import xarray as xr
```

```python
rcfile = "~/.polytopeapirc"
```

```python
kwargs =  {'address': 'polytope.lumi.apps.dte.destination-earth.eu'}
client = polytope.api.Client(**kwargs)
```

```python
client.list_collections()
```

```python
request = {
    'activity': 'ScenarioMIP',
    'class': 'd1',
    'dataset': 'climate-dt',
    'date': '20200102',
    'experiment': 'SSP3-7.0',
    'expver': '0001',
    'generation': '1',
    'levtype': 'sfc',
    'model': 'IFS-NEMO',
    'param': '134/165/166',
    'realization': '1',
    'resolution': 'standard',
    'stream': 'clte',
    'time': '0100', # '0100/0200/0300/0400/0500/0600'
    'type': 'fc'
}
```

```python
# Retrieve data
client.retrieve('destination-earth',  request, 'output_api.grib')
```

```python
dset = xr.open_dataset('output_api.grib', engine="cfgrib")
```
